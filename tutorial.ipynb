```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
```

```python
# g.add_node(1)
g.add_nodes_from([(4, {"color": "red"}), (5, {"color": "green"})])
```

```python
print(f"Nodes in the graph: {g.nodes}")
print(f"edges in the graph: {g.edges}")
```

```python
g.clear()
```

```python
print(g.nodes())
```

```python
g.add_edges_from([(1, 2), (1, 3), (2, 4)])
```

```python
g.add_node(1)   # No effect
g.add_edge(1, 2)
g.add_node("spam")
```

```python
g.add_nodes_from("filer")
g.add_edge(3, 'f')
```

```python
# checking adj list of a node
g.adj[2] # or list(G.neighbors(1))
```

```python
g.edges()
```

```python
g.degree[1]  # the number of edges incident to 1
```

```python
g.edges([2, 'f'])
```

```python
pos = nx.spring_layout(g)
pos
```

```python
nx.draw(g, pos, with_labels = True, node_color = 'orange')
```

```python
g.remove_node(2)
g.remove_nodes_from("filer")
list(g.nodes)
```

```python
pos = nx.spring_layout(g)
pos
```

```python
nx.draw(g, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=16)
plt.show()
```

```python
dg = nx.DiGraph()
```

```python
dg.add_edge(2, 1)   # adds the nodes in order 2, 1
dg.add_edge(1, 3)
dg.add_edge(2, 4)
dg.add_edge(1, 2)
```

```python
# assert list(dg.successors(2)) == [1, 4]
# assert list(dg.edges) == [(2, 1), (2, 4), (1, 3), (1, 2)]
```

```python
pos_dg = nx.spring_layout(dg)
pos_dg
```

```python
nx.draw(dg, pos_dg, with_labels = True)
```
